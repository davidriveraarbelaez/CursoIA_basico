```python
import numpy as np
import matplotlib.pyplot as plt

# Paso 1: Estandarización de Datos
np.random.seed(42)
# Generar datos de ejemplo en 2D para simplificar el ejemplo
X = np.random.rand(100, 2) * 10

# Calcular la media y la desviación estándar para cada dimensión
mean = np.mean(X, axis=0)
std_dev = np.std(X, axis=0)

# Estandarizar los datos: restar la media y dividir por la desviación estándar
X_standardized = (X - mean) / std_dev


```

```python
# Paso 2: Cálculo de la Matriz de Covarianza
cov_matrix = np.cov(X_standardized.T)

```

```python
# Paso 3: Cálculo de los Componentes Principales
# Calcular valores propios y vectores propios de la matriz de covarianza
eigenvalues, eigenvectors = np.linalg.eig(cov_matrix)

# Ordenar valores propios en orden descendente
idx = np.argsort(eigenvalues)[::-1]
eigenvalues = eigenvalues[idx]
eigenvectors = eigenvectors[:, idx]

```

```python
# Paso 4: Proyección de Datos en el Espacio de Componentes Principales
# Proyectar los datos estandarizados en el espacio de componentes principales
X_pca = X_standardized.dot(eigenvectors)

# Visualización de los Datos Transformados
plt.figure(figsize=(8, 6))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c='blue', alpha=0.5)
plt.xlabel('Componente Principal 1')
plt.ylabel('Componente Principal 2')
plt.title('Proyección en el Espacio de Componentes Principales')
plt.grid(True)
plt.show()
```
