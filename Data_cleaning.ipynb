```python
import pandas as pd
from sklearn.preprocessing import StandardScaler

# Cargar el dataset original
file_path = '/content/movie_metadata.csv'
df = pd.read_csv(file_path)
```

```python
# Selección de columnas numéricas relevantes
numeric_columns = [
    'num_critic_for_reviews', 'duration', 'director_facebook_likes',
    'actor_3_facebook_likes', 'actor_1_facebook_likes', 'gross',
    'num_user_for_reviews', 'budget', 'title_year', 'actor_2_facebook_likes',
    'imdb_score', 'aspect_ratio', 'movie_facebook_likes'
]

```

```python
# Filtrar el DataFrame para columnas numéricas seleccionadas
# Eliminar símbolos como $ en 'budget' y convertir a float
df_numeric = df[numeric_columns].replace('[\$,]', '', regex=True).astype(float)
df_numeric
```

```python
# Eliminar filas con valores nulos
df_numeric = df_numeric.dropna()
df_numeric
```

```python
# Eliminar filas duplicadas
df_numeric = df_numeric.drop_duplicates()
df_numeric
```

```python
# Estandarizar los datos
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df_numeric)
scaled_data
```

```python
# Descargar el dataset limpio generado

df_numeric.to_csv('movie_metadata_cleaned.csv', index=False)



```
