## Histogramas
Un histograma puede ser utilizado para mostrar la distribución de temperaturas diarias registradas en una región durante un año. Esto puede ayudar a identificar períodos de calor extremo o frío inusual, y permitir a los investigadores evaluar cómo estas variaciones pueden afectar los patrones de consumo energético o la salud de los cultivos.

```python
import numpy as np

import matplotlib.pyplot as plt

# Generar datos de ejemplo: temperaturas diarias durante un año
np.random.seed(0)
temperaturas = np.random.normal(loc=20, scale=5, size=365)

# Crear el histograma
plt.hist(temperaturas, bins=30, edgecolor='black')
plt.title('Distribución de Temperaturas Diarias')
plt.xlabel('Temperatura (°C)')
plt.ylabel('Frecuencia')
plt.grid(True)
plt.show()
```
