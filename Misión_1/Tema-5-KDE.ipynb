### Kernels de densidad Kernel Density Estimation (KDE)
Un gráfico de densidad puede representar la distribución de temperaturas registradas en una ciudad durante el transcurso de un año. Esto permite visualizar las variaciones estacionales y los períodos de calor extremo o frío, ayudando a los urbanistas y a las autoridades locales a tomar decisiones informadas sobre el manejo de recursos hídricos y la planificación urbana en función del clima.


```python
import seaborn as sns
import numpy as np

import matplotlib.pyplot as plt
```

```python
# Generar datos de ejemplo: temperaturas registradas durante un año
np.random.seed(0)
temperaturas = np.random.normal(loc=15, scale=10, size=365)  # Temperaturas con media de 15°C y desviación estándar de 10°C
```

```python
# Crear el gráfico de densidad
sns.kdeplot(temperaturas, fill=True)
plt.title('Distribución de Temperaturas Registradas en un Año')
plt.xlabel('Temperatura (°C)')
plt.ylabel('Densidad')
plt.show()
```
