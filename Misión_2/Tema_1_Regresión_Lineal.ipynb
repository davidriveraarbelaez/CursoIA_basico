```python
import matplotlib.pyplot as plt
import pandas as pd

# Carga de los datos
data = pd.read_csv('/content/product_ratings.csv')

# Selección de las variables
X = data['views'].values
y = data['rating'].values

# Visualización
plt.scatter(X, y)
plt.xlabel('Vistas')
plt.ylabel('Calificación')
plt.title('Relación entre Vistas y Calificación')
plt.show()
```

```python
import numpy as np
from scipy import stats

# Cálculo de la pendiente e intersección usando scipy
slope, intercept, r_value, p_value, std_err = stats.linregress(X, y)

# Generación de predicciones
y_pred = slope * X + intercept

# Visualización de la regresión
plt.scatter(X, y, label='Datos')
plt.plot(X, y_pred, color='red', label='Línea de Regresión')
plt.xlabel('Vistas')
plt.ylabel('Calificación')
plt.legend()
plt.title('Regresión Lineal usando Numpy y Scipy')
plt.show()

# Impresión de los resultados
print("Pendiente:", slope)
print("Intersección:", intercept)
print("R-cuadrado:", r_value**2)
```

```python
# Ejemplo: predecir el rating para 300 vistas
views_nuevo = 400
rating_pred_numpy_scipy = (slope * views_nuevo) + intercept
print("Predicción de rating (Numpy y Scipy) para", views_nuevo,   "vistas:", rating_pred_numpy_scipy)
```
