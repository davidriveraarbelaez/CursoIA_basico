```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import numpy as np
```

```python
"""
Se crea un array X con datos de ejemplo. 
Cada subarray representa un punto en un espacio bidimensional.
"""
# Datos de ejemplo
X = np.array([[2.5, 2.4], [0.5, 0.7], [2.2, 2.9], [1.9, 2.2], [3.1, 3.0], [2.3, 2.7], [2, 1.6], [1, 1.1], [1.5, 1.6], [1.1, 0.9]])
```

```python
# Estandarización de los datos
"""
Se estandarizan los datos para que cada característica tenga media cero y varianza uno. 
Esto es importante porque PCA es sensible a las escalas de las variables.
"""
X_std = StandardScaler().fit_transform(X) 
```

```python
# Aplicación de PCA
pca = PCA(n_components=2) # Seleccionamos 2 componentes principales, pero se pueden seleccionar menos
principalComponents = pca.fit_transform(X_std) # Se obtienen los componentes principales

print("Componentes principales:\n", principalComponents) # Se imprimen los componentes principales
```

```python
# Gráfica de los datos originales y los componentes principales

import matplotlib.pyplot as plt

plt.figure(figsize=(6, 6))
plt.scatter(X_std[:, 0], X_std[:, 1], color='blue', label='Datos originales')
plt.scatter(principalComponents[:, 0], principalComponents[:, 1], color='red', label='Componentes principales')
plt.legend()
plt.xlabel('X1')
plt.ylabel('X2')
plt.title('Datos originales y componentes principales')
plt.show()

# Varianza explicada por cada componente principal
print("Varianza explicada por cada componente principal:\n", pca.explained_variance_ratio_)
print("Varianza total explicada por los componentes principales:", sum(pca.explained_variance_ratio_)) # Varianza total explicada por los componentes principales
print("Varianza total de los datos originales:", np.var(X_std, axis=0).sum()) # Varianza total de los datos

# Se puede observar que la varianza total de los datos originales es igual a la varianza total explicada por los componentes principales.
# Esto es porque PCA es una técnica de reducción de dimensionalidad que busca conservar la mayor cantidad de varianza en los datos originales.
# En este caso, se conserva el 100% de la varianza.
```
